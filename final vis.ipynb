```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn import metrics
from sklearn.model_selection import train_test_split
df=pd.read_csv("populationdata.csv")
print(df.head())
X=df.iloc[1:26,0]
y=df.iloc[1:26,3]

labels=np.arange(1971,1996)

f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)

plt.xticks(X, labels)

plt.xlabel("year",fontdict={'c':'k','size':30})
plt.ylabel("population change",{'c':'k','size':30})
plt.title("UK population year wise change 1971-1995",c='black',size=40)
plt.plot(X,y,marker='o',markersize=12,markerfacecolor='red')
plt.show()
```

```python
data = pd.read_csv("populationdata.csv").values
feat  = np.asarray(data[:, :-1])
label = np.asarray(data[:, -1])
xtrain, xtest, ytrain, ytest = train_test_split(feat, label, test_size=0.3)


mdl = RandomForestClassifier()
mdl.fit(xtrain, ytrain)
y_pred    = mdl.predict(xtest)

print("##### Accuracy Parameters RandomForestClassifier #####\n")
print("Accuracy = ",metrics.accuracy_score(ytest, y_pred)*100)
print("Precision Score : ",metrics.precision_score(ytest, y_pred,average='micro'))
print("Recall Score : ",metrics.recall_score(ytest, y_pred,average='micro'))
print("F1 Score : ",metrics.f1_score(ytest, y_pred,average='micro'))
```
